# 파생변수 생성 (휴일여부, 요일, 휴일, 주말)

```python
sub18.info()
sub19.info()
```

```python
sub18 = sub18.astype({'사용일자':'datetime64'})
print(sub18.info())
```

```python
sub19 = sub19.astype({'사용일자':'datetime64'})
print(sub19.info())
```

```python
# 파생변수 생성
sub18['요일'] = sub18['사용일자'].dt.day_name()
```

```python
sub19['요일'] = sub19['사용일자'].dt.day_name()
```

```python
sub19
```

```python
# 요일 컬럼 추가
sub18['dayofweek'] = sub18['사용일자'].dt.dayofweek #요일 (월 = 0)

#0 1 2 3 4 5 6
#월화수목금토일
```

```python
# 요일 컬럼 추가
sub19['dayofweek'] = sub19['사용일자'].dt.dayofweek #요일 (월 = 0)

#0 1 2 3 4 5 6
#월화수목금토일
```

```python
sub19
```

```python
sub18_평일 = sub18[sub18['dayofweek']<=4]
sub18_평일
```

```python
sub19_평일 = sub19[sub19['dayofweek']<=4]
sub19_평일
```

```python
sub18_평일["데이"]=np.NaN
sub18_평일
```

```python
sub19_평일["데이"]=np.NaN
sub19_평일
```

```python
sub18_평일['데이'] = sub18_평일['데이'].fillna("1")
sub18_평일
```

```python
sub19_평일['데이'] = sub19_평일['데이'].fillna("1")
sub19_평일
```

```python
sub18_주말 = sub18[sub18['dayofweek']>=5]
sub18_주말
```

```python
sub19_주말 = sub19[sub19['dayofweek']>=5]
sub19_주말
```

```python
sub18_주말['휴일여부'] = sub18_주말['휴일여부'].fillna("Y")
sub18_주말
```

```python
sub19_주말['휴일여부'] = sub19_주말['휴일여부'].fillna("Y")
sub19_주말
```

```python
sub18_주말["데이"]=np.NaN
sub18_주말
```

```python
sub19_주말["데이"]=np.NaN
sub19_주말
```

```python
sub18_주말['데이'] = sub18_주말['데이'].fillna("0")
sub18_주말
```

```python
sub19_주말['데이'] = sub19_주말['데이'].fillna("0")
sub19_주말
```

```python
df18 = pd.concat([sub18_평일,sub18_주말])
df18
```

```python
df19 = pd.concat([sub19_평일,sub19_주말])
df19
```
